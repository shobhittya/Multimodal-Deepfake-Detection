```python
from tensorflow.keras.applications import MobileNetV2

# Load MobileNetV2 model for feature extraction (no classification head)
base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    pooling='avg',  # Global average pooling
    input_shape=(224, 224, 3)
)
```

```python

def extract_video_features(video_path, num_frames=10):
    """Extracts averaged EfficientNetB0 features from evenly spaced frames of a video."""
    if not os.path.isfile(video_path):
        print(f"Video file not found: {video_path}")
        return None

    print(f"Extracting audio from: {video_path}")
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if frame_count <= 0:
        return None

    frames = []
    print("Video Tested")
    # Select frame indices evenly across the video
    for i in np.linspace(0, frame_count-1, num_frames, dtype=int):
        cap.set(cv2.CAP_PROP_POS_FRAMES, i)
        ret, frame = cap.read()
        if not ret:
            continue
        # Convert BGR to RGB and resize
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (224, 224))
        frames.append(frame.astype(np.float32))

    cap.release()
    if not frames:
        return None

    # Preprocess frames and extract features batch-wise
    x = np.stack(frames)  # shape: (num_frames, 224, 224, 3)
    # Note: EfficientNet preprocess_input rescales pixels internally
    features = base_model.predict(x, verbose=0)
    video_feat = np.mean(features, axis=0)  # average pooling over frames
    return video_feat

# Example usage (will work if video files are present):
# sample_video = meta['full_path'].iloc[0]
# video_feat = extract_video_features(sample_video)
# print("Feature vector shape:", video_feat.shape)
```

```python

import io
def extract_audio_features(video_path, sr=16000, n_mels=64):
    """Extracts a simple mel-spectrogram feature vector from the video's audio track."""
    # Use ffmpeg to extract raw audio from video
    command = [
        'ffmpeg', '-y', '-i', video_path,
        '-ac', '1', '-ar', str(sr), '-f', 'wav', 'pipe:1'
    ]
    try:
        proc = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        out, err = proc.communicate()
        # Load audio from buffer
        y, sr_ret = librosa.load(io.BytesIO(out), sr=sr)
    except Exception as e:
        print(f"Error extracting audio: {e}")
        return None

    # Compute mel spectrogram
    S = librosa.feature.melspectrogram(y=y, sr=sr_ret, n_mels=n_mels)
    S_db = librosa.power_to_db(S, ref=np.max)
    # Aggregate over time by taking the mean of each Mel band
    mel_feat = np.mean(S_db, axis=1)
    return mel_feat

# Example usage:
# audio_feat = extract_audio_features(sample_video)
# print("Audio feature vector shape:", audio_feat.shape)
```

```python
import subprocess
import librosa
import numpy as np
import io

def extract_audio_features(video_path, sr=16000, n_mels=64):
    if not os.path.isfile(video_path):
        print(f"Video file not found: {video_path}")
        return None

    print(f"Extracting audio from: {video_path}")
    """
    Extracts a mel-spectrogram feature vector from the audio track of a video file.
    
    Parameters:
        video_path (str): Path to the input video file.
        sr (int): Target sampling rate.
        n_mels (int): Number of mel bands.
        
    Returns:
        np.ndarray or None: A vector of mean mel-spectrogram features, or None on failure.
    """
    command = [
        'ffmpeg', '-y', '-i', video_path,
        '-ac', '1', '-ar', str(sr), '-f', 'wav', 'pipe:1'
    ]

    try:
        proc = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        out, err = proc.communicate()

        if proc.returncode != 0:
            print(f"FFmpeg error: {err.decode()}")
            return None

        # Load audio from bytes buffer
        y, sr_ret = librosa.load(io.BytesIO(out), sr=sr)

        # Compute mel spectrogram
        S = librosa.feature.melspectrogram(y=y, sr=sr_ret, n_mels=n_mels)
        S_db = librosa.power_to_db(S, ref=np.max)

        # Aggregate over time by taking the mean of each Mel band
        mel_feat = np.mean(S_db, axis=1)
        return mel_feat

    except Exception as e:
        print(f"Error extracting audio: {e}")
        return None

# Example usage:
# features = extract_audio_features("example.mp4")
# if features is not None:
#     print("Audio feature shape:", features.shape)
```

```python
import numpy as np

X_video = np.load('X_video_all.npy')
X_audio = np.load('X_audio_all.npy')
y= np.load('y_all.npy')
```

```python
from tensorflow.keras.layers import Input, Dense, BatchNormalization, Dropout, LeakyReLU, concatenate
from tensorflow.keras.models import Model

# Define inputs
video_input = Input(shape=(1280,), name='video_input')  # from Xception features
audio_input = Input(shape=(64,), name='audio_input')    # from MFCC 1D-CNN

# Video branch
v = Dense(512)(video_input)
v = BatchNormalization()(v)
v = LeakyReLU()(v)
v = Dropout(0.3)(v)

v = Dense(256)(v)
v = BatchNormalization()(v)
v = LeakyReLU()(v)
v = Dropout(0.3)(v)

# Audio branch
a = Dense(128)(audio_input)
a = BatchNormalization()(a)
a = LeakyReLU()(a)
a = Dropout(0.3)(a)

a = Dense(64)(a)
a = BatchNormalization()(a)
a = LeakyReLU()(a)
a = Dropout(0.3)(a)

# Merge branches
combined = concatenate([v, a])
x = Dense(128)(combined)
x = BatchNormalization()(x)
x = LeakyReLU()(x)
x = Dropout(0.4)(x)

# Output layer
output = Dense(len(classes), activation='softmax')(x)  # use 'sigmoid' if binary

# Build model
model = Model(inputs=[video_input, audio_input], outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
from sklearn.model_selection import train_test_split
Xv_train, Xv_val, Xa_train, Xa_val, y_train, y_val = train_test_split(
    X_video, X_audio, y, test_size=0.3, random_state=42)

# Train the model
history = model.fit(
    x=[Xv_train, Xa_train], y=y_train,
    validation_data=([Xv_val, Xa_val], y_val),
    epochs=200, batch_size=10
)
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss
plt.figure(figsize=(8,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Training vs Validation Loss')
plt.savefig('MobileNet_loss_plot.png')  # 🔸 Save the loss graph
plt.show()

# Plot training & validation accuracy
plt.figure(figsize=(8,4))
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training vs Validation Accuracy')
plt.savefig('MobileNet_accuracy_plot.png')  # 🔸 Save the accuracy graph
plt.show()
```
